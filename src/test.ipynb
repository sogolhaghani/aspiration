```python
from torch.utils.data import Dataset
import pandas as pd
import torch
from torch.utils.data import DataLoader
from transformers import AutoTokenizer,AutoModelForTokenClassification
from transformers import pipeline
```

```python
class CustomTextDataset(Dataset):
    def __init__(self, df, transform=None):
        self.df = df
        self.transform = transform

    def __len__(self):
        return len(self.df)
        
    def __getitem__(self, idx):
        text = self.df.loc[idx][0]
        sample = {"text": text, "label": 0}
        return text
```

```python
_read_directory = './aspiration/'
_doc = ['Thinking fast and Slow',
'Theories of Personality',
'WHY MEN LOVE Bitches']

_raw_data = pd.read_csv(_read_directory + _doc[0] + '.txt', delimiter = "\t", header=None)
dataset = CustomTextDataset(_raw_data)
train_dataloader = DataLoader(dataset, batch_size=64, shuffle=True)

tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")

nlp = pipeline("ner", model=model, tokenizer=tokenizer)
for batch in train_dataloader:
  ner_results = nlp(batch)
  print(ner_results)
```
